# Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

---

## Example 1:

```
Input: strs = ["flower","flow","flight"]
Output: "fl"
```

---

## Example 2:

```
Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.
```

---

## Constraints:

```
1 <= strs.length <= 200
0 <= strs[i].length <= 200
strs[i] consists of only lowercase English letters if it is non-empty.
```

---

````python
class Solution(object):
    def longestCommonPrefix(self, strs):
        if not strs:
            return ""

        # Scan through the length of the FIRST string
        for i in range(len(strs[0])):
            char = strs[0][i]

            # Check this character against ALL other strings
            for j in range(1, len(strs)):
                # CRITICAL CHECKS:
                # 1. Is the index i out of bounds for string j? (If strs[j] is shorter)
                # 2. Does the character not match?
                if i == len(strs[j]) or strs[j][i] != char:
                    # If either is true, the prefix ends exactly here.
                    # Return everything up to (but not including) i.
                    return strs[0][:i]

        # If we finish the loop without returning, it means the
        # entire first string is the common prefix.
        return strs[0]
````
