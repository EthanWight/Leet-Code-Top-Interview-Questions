# Reverse Integer

Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

---

## Example 1:

```
Input: x = 123
Output: 321
```

---

## Example 2:

```
Input: x = -123
Output: -321
```

---

## Example 3:

```
Input: x = 120
Output: 21
```

---

## Constraints:

```
-231 <= x <= 231 - 1
```

---

````python
class Solution(object):
    def reverse(self, x):
        negative = False
        if x < 0:
            negative = True
            x *= -1
        intStr = str(x)
        reversedInt = intStr[::-1]
        newInt = int(reversedInt)
        if newInt > pow(2,31) - 1:
            return 0
        if negative:
            newInt *= -1
        return newInt
````
