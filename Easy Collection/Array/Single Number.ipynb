# Single Number

Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

---

## Example 1:

```
Input: nums = [2,2,1]
Output: 1
```

---

## Example 2:

```
Input: nums = [4,1,2,1,2]
Output: 4
```

---
## Example 3:

```
Input: nums = [1]
Output: 1
```

---

## Constraints:

```
1 <= nums.length <= 3 * 104
-3 * 104 <= nums[i] <= 3 * 104
Each element in the array appears twice except for one element which appears only once.
```

---

````python
class Solution(object):
    def singleNumber(self, nums):
        #Declare num dictionary
        numDict = {}
        #for num in nums list
        for num in nums:
            #if num is already in dictionary increment
            if num in numDict:
                numDict[num] += 1
            #if num is not in dictionary set its initial value to 1
            else:
                numDict[num] = 1

        #for each value pair check if count is equal to 1
        for num, count in numDict.items():
            #if count is equal to 1 element appears once
            if count == 1:
                return num
````
