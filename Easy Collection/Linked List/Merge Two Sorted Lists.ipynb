# Merge Two Sorted Lists

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

---

## Example 1:

```
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
```

---

## Example 2:

```
Input: list1 = [], list2 = []
Output: []
```

---

## Example 3:

```
Input: list1 = [], list2 = [0]
Output: [0]
```

---

## Constraints:

```
The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.
```

---

````python
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        # 1. Create a dummy head to start the list
        dummy = ListNode(0)
        curr = dummy

        # 2. Iterate while both lists have nodes
        while list1 and list2:
            if list1.val < list2.val:
                curr.next = list1
                list1 = list1.next
            else:
                curr.next = list2
                list2 = list2.next
            # Move our 'curr' pointer forward
            curr = curr.next

        # 3. If one list is longer than the other,
        # just attach the remainder to the end.
        curr.next = list1 if list1 else list2

        # 4. Return dummy.next (the actual start of the merged list)
        return dummy.next
````
